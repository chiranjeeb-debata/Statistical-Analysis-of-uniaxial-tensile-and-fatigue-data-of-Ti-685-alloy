# **Extract Text from PDFs**

```python
def get_pdf_text(pdf_docs):

    text = ""

    for pdf in pdf_docs:

        pdf_reader = PdfReader(pdf)

        for page_num, page in enumerate(pdf_reader.pages):

            page_text = page.extract_text()

            text += f"\n\nDOCUMENT: {pdf}\n"
            text += f"PAGE: {page_num + 1}\n\n"

            text += page_text

    return text

raw_text = get_pdf_text(uploaded.keys())

print(raw_text[:1000])
```

# **Word Embedding**

Split Text into Chunks

```python
def get_text_chunks(text):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=50
    )

    chunks = text_splitter.split_text(text)

    return chunks


text_chunks = get_text_chunks(raw_text)

print("Number of chunks:", len(text_chunks))
```

```python
def get_vector_store(text_chunks):

    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    vector_store = FAISS.from_texts(
        text_chunks,
        embedding=embeddings
    )

    vector_store.save_local("faiss_index")


get_vector_store(text_chunks)

print("Vector DB created successfully")
```

# **Create Prompt Template**

```python
prompt_template = """
Answer the question as detailed as possible using the provided context.

If the answer is not in the provided context, say:
"answer is not available in the context"

Context:
{context}

Question:
{question}

Answer:
"""


prompt = PromptTemplate(
    template=prompt_template,
    input_variables=["context", "question"]
)
```

# **Load Gemini Model**

```python
def get_conversational_chain():

    model = ChatGoogleGenerativeAI(
        model="gemini-flash-latest",
        temperature=0.3
    )

    # Define a custom callable object that mimics the behavior of a LangChain 'stuff' chain
    # This will take 'input_documents' and 'question' as input,
    # format them with the 'prompt', and invoke the model.
    class CustomQAChain:
        def __init__(self, model, prompt):
            self.model = model
            self.prompt = prompt

        def __call__(self, inputs, **kwargs):
            input_documents = inputs["input_documents"]
            question = inputs["question"]

            # Combine document page content into a single string for the 'context'
            context = "\n\n".join([doc.page_content for doc in input_documents])

            # Format the prompt using the combined context and the question
            formatted_prompt = self.prompt.format(context=context, question=question)

            # Invoke the model with the formatted prompt
            response_content = self.model.invoke(formatted_prompt).content

            # Return the response in the dictionary format expected by the 'user_input' function
            return {"output_text": response_content}

    # Return an instance of our custom chain
    return CustomQAChain(model, prompt)
```

# **Ask Questions**

```python
def user_input(user_question):

    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    new_db = FAISS.load_local(
        "faiss_index",
        embeddings,
        allow_dangerous_deserialization=True
    )

    docs = new_db.similarity_search(user_question)

    chain = get_conversational_chain()

    response = chain(
        {
            "input_documents": docs,
            "question": user_question
        },
        return_only_outputs=True
    )

    print("\nReply:\n")
    print(response["output_text"])
```

```python
%%writefile app.py

import streamlit as st
import os

from pypdf import PdfReader
from langchain_text_splitters import RecursiveCharacterTextSplitter

from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

from langchain_google_genai import ChatGoogleGenerativeAI


os.environ["GOOGLE_API_KEY"] = "YOUR_API_KEY"


st.title("AI Research Paper Assistant")


uploaded_files = st.file_uploader(
    "Upload PDFs",
    accept_multiple_files=True
)


def get_pdf_text(pdf_docs):

    text = ""

    for pdf in pdf_docs:

        pdf_reader = PdfReader(pdf)

        for page in pdf_reader.pages:
            text += page.extract_text()

    return text


def get_text_chunks(text):

    splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=50
    )

    return splitter.split_text(text)


def create_vector_db(chunks):

    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    vector_store = FAISS.from_texts(
        chunks,
        embedding=embeddings
    )

    vector_store.save_local("faiss_index")


def ask_question(question):

    embeddings = HuggingFaceEmbeddings(
        model_name="sentence-transformers/all-MiniLM-L6-v2"
    )

    db = FAISS.load_local(
        "faiss_index",
        embeddings,
        allow_dangerous_deserialization=True
    )

    docs = db.similarity_search(question, k=2)

    context = "\n\n".join(
        [doc.page_content for doc in docs]
    )

    model = ChatGoogleGenerativeAI(
        model="gemini-2.0-flash",
        temperature=0.3
    )

    prompt = f"""
    Answer using the provided context.

    Context:
    {context}

    Question:
    {question}

    Answer:
    """

    response = model.invoke(prompt)

    return response.content


if uploaded_files:

    raw_text = get_pdf_text(uploaded_files)

    chunks = get_text_chunks(raw_text)

    create_vector_db(chunks)

    st.success("Documents processed successfully")


user_question = st.text_input("Ask a question")

if user_question:

    answer = ask_question(user_question)

    st.write(answer)
```
